```python
import pathlib
import multiprocessing
import functools

import seaborn as sns

import main
import gp
```

```python
data_path = pathlib.Path('./containerfs/tmp/cetdl1772small.dat')
print(f'Data exists: {data_path.is_file()}')
```

```python
training_data = main.parse_data(data_path)
RUNS = 20
GENERATIONS = 20
```

```python
def do_run(generations, lambda_, i):
    print(f'RUN {i}')
    gpobj = gp.GP(lambda_, training_data)
    return gpobj.run(generations=generations).fitness
```

```python
generations = 20
lambda_ = 20
with multiprocessing.Pool() as pool:
    func = functools.partial(do_run, generations, lambda_)
    gp1_fitnesses = pool.map(func, range(RUNS))

display(gp1_fitnesses)
```

```python
generations = 50
lambda_ = 20
with multiprocessing.Pool() as pool:
    func = functools.partial(do_run, generations, lambda_)
    gp2_fitnesses = pool.map(func, range(RUNS))

display(gp2_fitnesses)
```

```python
generations = 20
lambda_ = 50
with multiprocessing.Pool() as pool:
    func = functools.partial(do_run, generations, lambda_)
    gp3_fitnesses = pool.map(func, range(RUNS))

display(gp3_fitnesses)
```

```python
x = [
    gp1_fitnesses,
    gp2_fitnesses,
    gp3_fitnesses
]
y = [
    'gen=20, lambda=20',
    'gen=50, lambda=20',
    'gen=20, lambda=50'
]
```

```python
sns.boxplot(x, y)
```
